# Prepare Data to Feed Model

```python
# Training and testing dataset (inputs)
# Drop the target column 'fake'
X_train = instagram_df_train.drop(columns = ['fake'])
X_test = instagram_df_test.drop(columns = ['fake'])
```

```python
X_train
```

```python
X_test
```

```python
# Training and testing dataset (Outputs)
# Set target column 'fake' as output
y_train = instagram_df_train['fake']
y_test = instagram_df_test['fake']
```

```python
y_train
```

```python
y_test
```

```python
# Scale the data before training the model
# Normalizing will help prevent one feature from becoming dominant over others
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler_x = StandardScaler()
X_train = scaler_x.fit_transform(X_train)
X_test = scaler_x.transform(X_test)
```

```python
# Convert single column into two categorical columns (2 classes)
y_train = tf.keras.utils.to_categorical(y_train, num_classes = 2)
y_test = tf.keras.utils.to_categorical(y_test, num_classes = 2)
```

```python
y_train
```

```python
y_test
```

```python
# Print out the shapes of the training and testing datasets
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

Determine percentage of training and testing data.

```python
Training_data = len(X_train) / (len(X_train) + len(X_test)) * 100
Training_data # 83% of data has been allocated for training
```

```python
Test_data = len(X_test) / (len(X_train) + len(X_test)) * 100
Test_data # 17% of data has been allocated for testing
```

# Build and Train a Simple Deel Learning Model

```python
# Build an artificial neural network
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

# Tweak parameters to achieve high accuracy
model = Sequential()
model.add(Dense(50, input_dim = 11, activation = 'relu')) # add layers
model.add(Dense(150, activation = 'relu'))
model.add(Dense(25, activation = 'relu'))

model.add(Dropout(0.3)) # drop a couple neurons to prevent overfitting

model.add(Dense(2, activation = 'softmax')) # activation function
model.summary()
```

```python
# Compile model
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
# Fit the model
# Epochs refers to a complete cycle of the training process
epochs_hist = model.fit(X_train, y_train, epochs = 15,  verbose = 1, validation_split = 0.1)
# result 92% accuracy
```

# Assess Performance of Trained Model

```python
print(epochs_hist.history.keys())
```

```python
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])

plt.title('Model Loss Progression During Training/Validation')
plt.ylabel('Training and Validation Losses')
plt.xlabel('Epoch Number')
plt.legend(['Training Loss', 'Validation Loss'])

# both training loss and validation loss decreased
# our model does not overfit the training data
```

```python
# Predict
predicted = model.predict(X_test)
```

```python
predicted_value = []
test = []
for i in predicted:
    predicted_value.append(np.argmax(i))

for i in y_test:
    test.append(np.argmax(i))
```

```python
print(classification_report(test, predicted_value))
```

```python
# Generate confusion matrix
plt.figure(figsize=(10, 10))
cm = confusion_matrix(test, predicted_value)
sns.heatmap(cm, annot = True)
# correctly classfied 54 and 53 samples
# incorrectly classfied 6 and 7 samples
```
